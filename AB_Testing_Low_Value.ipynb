## ****Low Value Segment Analysis****

### ***Verify Random Assignment for test vs control group for all the pre-test variables***

*Base vs Control*

Other ways to check random assignment:
- Check the method used: is a random generator/algorithm used?
- Visual - similar shape and spreads

```python
#low-value base
low_value_segment[low_value_segment['base'] == 1]['group'].value_counts()
```

```python
low_base = low_value_segment[low_value_segment['base'] == 1]
```

```python
#low-value control
low_value_segment[low_value_segment['base'] == 0]['group'].value_counts()
```

```python
low_control = low_value_segment[low_value_segment['base'] == 0]
```

```python
#Verify Random Assignment for test vs control group for all the pre-test variables 

from scipy import stats

#1.totdol - total spend in past 12 months
#Null hypothesis: there is no relationship between group assignment (test/control) and spending patterns.
t_stat, p_value = stats.ttest_ind(low_control['totdol'], low_base['totdol'], equal_var=True)  # set equal_var=False for Welch’s t-test
print("p-value for totdol:", p_value)

#2.last - number of days since last purchase
t_stat, p_value = stats.ttest_ind(low_control['last'], low_base['last'], equal_var=True)  # set equal_var=False for Welch’s t-test
print("p-value for last:", p_value)

#3. numpurch - number of purchases in past 12 months
t_stat, p_value = stats.ttest_ind(low_control['numpurch'], low_base['numpurch'], equal_var = True)
print("p-value for numpurch: ", p_value)

#4. gender
t_stat, p_value = stats.ttest_ind(low_control['gender'], low_base['gender'], equal_var = True)
print("p-value for gender: ", p_value)

#5. age
t_stat, p_value = stats.ttest_ind(low_control['age'], low_base['age'], equal_var = True)
print("p-value for age: ", p_value)
```

***In conclusion, P values for totdol, last, numpurch, and gender are all higher than 0.5 means that the null hypothesis cannot be rejected, which means randomization was achieved in the low-value segment between test and control groups. However, in terms of age, p value being 0.04, which is lower than 0.05. Null hypothesis is being rejected, which means randomization was NOT achieved in the low-value segment in terms of age. It means there is not enough evidence to support that the test and control groups are sufficiently randomized by age. Any conclusion of coupon use based on coupon use for low-value segment could be off.***

### ***Analyze Purchase Behavior***

*Base vs Control*

```python
low_value_segment['group'].value_counts()
```

```python
low_value_segment.groupby('group')['purchase'].mean()*100
```

```python
summary1 = (
    low_value_segment.groupby('group').agg(
        customers= ('purchase', 'count'),  # number of observations
        purchase_rate= ('purchase', lambda x: x.mean() * 100),   # % of 1s if binary 
        redeem_rate = ('redeem', lambda x: x.mean() * 100),
        ARPC = ('amount', 'mean') # Average Revenue per Customer   
    ))

print(summary1)
```

```python
#low_value buyer df
low_value_buyer = low_value_segment[low_value_segment['purchase'] == 1]
```

```python
summary2 = (
    low_value_buyer.groupby('group')['amount'].mean().reset_index(name='ARPB')
)

print(summary2)
```

```python
summary = summary1.merge(summary2, on='group', how='left')

print(summary.round(2))
```

```python
#Calculate p value for purchase_rate
#Ho: Base coupon doesn't affect the purchase rate for the low-value segment
#Ha: Base coupon significantly affect the purchase rate for the low-value segment

p_purchase= stats.ttest_ind(low_control['purchase'], low_base['purchase'], equal_var=True).pvalue # set equal_var=False for Welch’s t-test
print("p-value for purchase:", p_purchase)
```

```python
summary['p_purchase_rate'] = summary['group'].map({
    'CG_L': None,
    'T1_L': p_purchase
})

print(summary.round(4))
```

The p-value for purchase rate is 0.001, less than 0.05, so we can make the conclusion that base coupon significantly affect the purchase rate for the low-value segment from 1% to 2.47%.

```python
#Calculate p value for amount (ARPC)
p_ARPC= stats.ttest_ind(low_control['amount'], low_base['amount'], equal_var=True).pvalue  # set equal_var=False for Welch’s t-test
print("p-value for ARPC:", p_ARPC)
```

```python
summary['p_ARPC'] = summary['group'].map({
    'CG_L': None,
    'T1_L': p_ARPC
})

print(summary.round(4))
```

The p-value for APRC is 0.008, less than 0.05, so we can make the conclusion that the base coupon significantly increases the revenue for the low-value segment from 0.26 to 0.71.

```python
#Calculate p value for ARPB

low_buyer_control = low_value_segment[(low_value_segment['purchase'] == 1) & (low_value_segment['base'] == 0)]
low_buyer_base = low_value_segment[(low_value_segment['purchase'] == 1) & (low_value_segment['base'] == 1)]
p_ARPB = stats.ttest_ind(low_buyer_control['amount'], low_buyer_base['amount'], equal_var = True).pvalue

print("p value for ARPB:", p_ARPB)
```

```python
summary['p_ARPB'] = summary['group'].map({
    'CG_L': None,
    'T1_L': p_ARPB
})

print(summary.round(4))
```

The p-value for ARPB is 0.52, higher than 0.05, we cannot reject null hypothesis, meaning that there is no statistically significant impact on average revenue per buyer for low-value segment when they receive base coupon.

***Conclusion: For low-value segment, the base coupon is effective in increasing purchase rate and purchase amount, but not effective in increasing the purchase amount among the buyers. In other words, the base coupon increases the revenue among the low-value segment because of the increase in the pruchase rate, but doesn't make buyers spend more.***

### ***Linear Regression***

After conducting  A/B testing, the purpose of creating a linear regression model is to gain deeper insights and quantify the relationship between variables.

1- Adjust for Covariates (control variables): a linear model allows you to control for these covariates (eg. age, location, gender), isolating the true effect of the treatment.

2- Estimate the Effect Size: Gives you an estimated lift (β coefficient) for the test group.

3- Predictive Power: Linear models can be used to predict future outcomes based on the treatment and other variables.

```python
import statsmodels.api as sm
```

```python
#For low-value segment, is there a relationsihp between receiving the base coupon and purchase amount?
#base vs amount

X = low_value_segment[['base']]
y = low_value_segment['amount']

# Add constant for intercept
X = sm.add_constant(X)

# Fit OLS model
linear_low= sm.OLS(y, X).fit()

# Print summary
print(linear_low.summary())
```

1. Test the overall significance of the regression model compared to a model with no independent varilabes.
   The P-value for the model is 0.000835 --> Reject null, meaning that the regression model could be valuable.
2. Assess how good the regression model is.
   Adjusted R-squared is 0.001 --> The regression model can explain about 0.1% of the information in the dependent variable.
3. Test the significance of each individual coefficient
   the coef is 0.4588, and p value is 0.001 --> reject null, meaning the base coupon is a significant predictor of purchase amount resulting in a lift of 0.46, which matches the difference between low_base and low_control in ARPC.
4. Make predictions.
   

***Conclusion: the base coupon increases the revenue by 0.46 on average, compared to control.***

```python
# Make predictions (like R's predict(..., type='response'))
amount_hat = linear_low.predict(X)

# Summary statistics (like R's summary(amount_hat))
print(pd.Series(amount_hat).describe())
```
