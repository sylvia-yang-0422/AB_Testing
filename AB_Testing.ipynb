**Verify Random Assignment for test vs control group for all the pre-test variables**

*Low_Value Segment*

```python
#Verify Random Assignment for test vs control group for all the pre-test variables 
from scipy import stats

#Low_value segment
#1.totdol - total spend in past 12 months
#Null hypothesis: there is no relationship between group assignment (test/control) and spending patterns.
control_totdol = low_value_segment[low_value_segment['base'] == 0]['totdol']
test_totdol = low_value_segment[low_value_segment['base'] == 1]['totdol']

t_stat, p_value = stats.ttest_ind(control_totdol, test_totdol, equal_var=True)  # set equal_var=False for Welch’s t-test
print("t-statistic:", t_stat)
print("p-value:", p_value)
```

P value bigger than 0.5 means that the null hypothesis cannot be rejected, which means randomization was achieved in the low-value segment between test and control groups in terms of totdor.

```python
#2. last - number of days since last purchase
control_last = low_value_segment[low_value_segment['base'] == 0]['last']
test_last = low_value_segment[low_value_segment['base'] == 1]['last']

t_stat, p_value = stats.ttest_ind(control_last, test_last, equal_var = True)
print('t-statistics:', t_stat)
print('p-value:', p_value)
```

P value is 0.159, which is bigger than 0.5, we cannot reject null hypothesis, which means randomization was achieved in the low-value segment between test and control groups in terms of number of days since last purchase.

```python
#3. numpurch - number of purchases in past 12 months
control_numpurch = low_value_segment[low_value_segment['base'] == 0]['numpurch']
test_numpurch = low_value_segment[low_value_segment['base'] == 1]['numpurch']

t_stat, p_value = stats.ttest_ind(control_numpurch, test_numpurch, equal_var = True)
print('t-statistics:', t_stat)
print('p-value:', p_value)
```

P value is 0.195, which is bigger than 0.05. We cannot reject the null hypothesis, which means randomization was achieved in the low-value segment between test and control groups in terms of number of purchases in the past 12 months.

```python
#4. gender
control_gender = low_value_segment[low_value_segment['base'] == 0]['gender']
test_gender = low_value_segment[low_value_segment['base'] == 1]['gender']

t_stat, p_value = stats.ttest_ind(control_gender, test_gender, equal_var = True)
print('t-statistic:', t_stat)
print('p-value:', p_value)
```

P value is 0.54, which is higher than 0.05. We cannot reject null hypothesis, which means randomization was achieved in the low-value segment between test and control groups in terms of gender.

```python
#5. age
control_age = low_value_segment[low_value_segment['base'] == 0]['age']
test_age = low_value_segment[low_value_segment['base'] == 1]['age']

t_stat, p_value = stats.ttest_ind(control_age, test_age, equal_var = True)
print('t-statistic:', t_stat)
print('p-value:', p_value)
```

? P value is 0.04, which is lower than 0.05. We are rejecting null hypothesis, which means randomization was NOT achieved in the low-value segment between test and control groups in terms of age.

**Linear Regression**

```python
import statsmodels.api as sm
```

```python
#Low_value segment
#Is there a relationsihp between receiving the base coupon and purchase amount?
#base vs amount

X = low_value_segment[['base']]
y = low_value_segment['amount']

# Add constant for intercept
X = sm.add_constant(X)

# Fit OLS model
linear_low= sm.OLS(y, X).fit()

# Print summary
print(linear_low.summary())
```

1. Test the overall significance of the regression model compared to a model with no independent varilabes.
   The P-value for the model is 0.000835 --> Reject null, meaning that the regression model could be valuable.
2. Assess how good the regression model is.
   Adjusted R^2 is 0.001 --> The regression model can explain about 0.1% of the information in the dependent variable.
3. Test the significance of each individual coefficient
   the coef is 0.4588, and p value is 0.001 --> reject null, meaning the base coupon is a significant predictor of purchase amount resulting in a lift of $0.46.
4. Make predictions.
   

```python
# Make predictions (like R's predict(..., type='response'))
amount_hat = linear_low.predict(X)

# Summary statistics (like R's summary(amount_hat))
print(pd.Series(amount_hat).describe())
```

```python
#High_value segment
#Is there a relationsihp between receiving the base coupon and purchase amount?
#base vs amount

X = high_value_segment[['base', 'better']]
y = high_value_segment['amount']

# Add constant for intercept
X = sm.add_constant(X)

# Fit OLS model
linear_high= sm.OLS(y, X).fit()

# Print summary
print(linear_high.summary())
```

1. Test the overall significance of the regression model compared to a model with no independent varilabes.
   The P-value for the model is 0.000928 --> Reject null, meaning that the regression model is significant.
2. Assess how good the regression model is.
   Adjusted R^2 is 0.002 --> The regression model can explain about 0.2% of the information in the dependent variable.
3. Test the significance of each individual coefficient
   the coef for base coupon is 1.36 and for better coupon is 1.97, and p values are 0.016 and 0.018 -> reject null, meaning both coupon choices are effective in increasing purchase amount(a lift of 0.36 and 1.97) in the high-value sgment and the better coupon is more effective than the base coupon.
4. Make predictions.

```python
# Make predictions (like R's predict(..., type='response'))
amount_hat_high = linear_high.predict(X)

# Summary statistics (like R's summary(amount_hat))
print(pd.Series(amount_hat).describe())
```
